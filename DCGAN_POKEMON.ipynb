```python
folders = []
get_images_path=[]

def read_dir():
    # Read folder of different images
    path = 'C:\\Users\\I353174\\Desktop\\DCGAN\\dataset'
    for root, directories, files in os.walk(path):
        for get_folder in directories:
             folders.append(os.path.join(root, get_folder))
    # Get all images path           
    for root, directories, files in os.walk(path):
          for file in files:
                if '.jpg' in file:
                    get_images_path.append(os.path.join(root, file))
```

```python
# Method to copy all images in one folder
def copy_images():
    create_dir = Path('pokemon_dataset')
    create_dir.mkdir(exist_ok=True)
    dst = 'C:\\Users\\I353174\\Desktop\\DCGAN\\pokemon_dataset'
    if os.listdir(dst) == []: 
        for src in get_images_path:
            shutil.copy(src, dst)
    else :
        print("Directory is not empty")
```

```python
''' Resize the width and height of image into common scale, vectorize pixel of the image and store as matrix '''
image_mat=[]
real_images=[]
def image_matrix():
    new_width  = 128
    new_height = 128
    if(image_mat==[] and real_images==[]):
        for x in images:
            img=Image.open('pokemon_dataset\\'+x)
            img = img.resize((new_width, new_height), Image.ANTIALIAS)
            img_arr=np.array(img) 
            if(img_arr.ndim == 3):
                real_images.append(img_arr)
                height,width,channel = img_arr.shape
                img_size =  height*width*channel 
                img_1D_vector = img_arr.reshape(img_size)
                image_mat.append(img_1D_vector)
                # image_shape.append(img_arr.shape) 
    else:
        print("Image matrix is not empty")
```

```python
original_img=[]
def reformat_image():
    read_dir()
    copy_images()
    image_matrix() 
    imgs=np.asarray(image_mat[:6002])
    for x in range(len(imgs)):
        if(imgs[x].shape!=(65536,)):
             original_img.append(imgs[x].reshape(128,128,3))    
```

```python
def generator():
    
    generator = Sequential()
#     generator.add(Dense(units=4 * 4 * 128, input_dim=100 ))
#     generator.add(BatchNormalization(momentum=0.5))
#     generator.add(Activation('relu'))
    
    generator.add(Dense(units=8* 8 * 256, input_dim=100 ))
    generator.add(Reshape(target_shape=(8, 8, 256)))
    generator.add(BatchNormalization(momentum=0.5))
    generator.add(Activation('relu'))
    generator.add(UpSampling2D(size=(2, 2)))
    generator.add(BatchNormalization(momentum=0.5))
    generator.add(Activation('relu'))
    generator.add(UpSampling2D(size=(2, 2)))
    generator.add(BatchNormalization(momentum=0.5))
    generator.add(Activation('relu')) 
    generator.add(Conv2D(filters=128, kernel_size=(5, 5), strides=(2, 2), padding='same', data_format='channels_last'))
    generator.add(BatchNormalization(momentum=0.5))
    generator.add(Activation('relu')) 
    generator.add(UpSampling2D(size=(2, 2)))
    generator.add(BatchNormalization(momentum=0.5))
    generator.add(Activation('relu'))   
    generator.add(Conv2D(filters=64, kernel_size=(5, 5), strides=(2, 2), padding='same', data_format='channels_last'))
    generator.add(BatchNormalization(momentum=0.5))
    generator.add(Activation('relu'))  
    generator.add(UpSampling2D(size=(4, 4)))
    generator.add(BatchNormalization(momentum=0.5))
    generator.add(Activation('relu'))   
    generator.add(Conv2D(filters=34, kernel_size=(5, 5), strides=(2, 2), padding='same', data_format='channels_last'))
    generator.add(BatchNormalization(momentum=0.5))
    generator.add(Activation('relu'))   
    generator.add(UpSampling2D(size=(8, 8)))
    generator.add(BatchNormalization(momentum=0.5))
    generator.add(Activation('relu'))   
    generator.add(Conv2D(filters=3, kernel_size=(5, 5), strides=(2, 2), padding='same', data_format='channels_last'))
    generator.add(Activation('tanh'))
    
    return generator
```

```python
def discriminator():
    
    discriminator = Sequential()
    
    discriminator.add(Conv2D(filters=64, kernel_size=(5, 5), strides=(2, 2), padding='same', data_format='channels_last', input_shape=(128,128,3)))
    discriminator.add(BatchNormalization(momentum=0.5))
    discriminator.add(LeakyReLU(0.2)) 
    discriminator.add(Conv2D(filters=64, kernel_size=(5, 5), strides=(2, 2), padding='same', data_format='channels_last'))
    discriminator.add(BatchNormalization(momentum=0.5))
    discriminator.add(LeakyReLU(0.2))  
    discriminator.add(BatchNormalization())
    discriminator.add(MaxPooling2D(strides=(2,2)))
    discriminator.add(Dropout(0.25)) 
    discriminator.add(Conv2D(filters=128, kernel_size=(5, 5),strides=(2, 2), padding='same', data_format='channels_last'))
    discriminator.add(BatchNormalization(momentum=0.5))
    discriminator.add(LeakyReLU(0.2))
    discriminator.add(MaxPooling2D(strides=(2,2)))
    discriminator.add(Flatten())
    discriminator.add(Dense(512))
    discriminator.add(LeakyReLU(0.2))
    discriminator.add(Dropout(0.25))
    discriminator.add(Dense(1024))
    discriminator.add(Dense(1))
    discriminator.add(Activation('sigmoid'))
    return discriminator
```

```python
def generator_discriminator(gen , disc):
    model = Sequential()
    model.add(gen)
    disc.trainable = False
    model.add(disc)
    return model
```

```python
# Method to build genrative adversarial network and train both the discriminators and generators
def train_gan(batch_size, epochs):
#     gen = generator()
#     disc = discriminator()
     
    gen =  load_model("models\\generator_epoch6(39).hdf5")
    disc =  load_model("models\\discriminator_epoch6(39).hdf5")
    gan = generator_discriminator(gen , disc)

    d_optim = SGD(lr=0.0005, momentum=0.9, nesterov=True)
    g_optim = SGD(lr=0.0005, momentum=0.9, nesterov=True)
    
    gen.compile(loss='binary_crossentropy', optimizer="SGD")
    print("Generator initialized .................")
    gan.compile(loss='binary_crossentropy', optimizer=g_optim)
    print("Adversarial initialized .................")
    disc.trainable = True
    disc.compile(loss='binary_crossentropy', optimizer=d_optim)
    print("Discriminator initialized .................")
    number_of_batches = int(len(result)/batch_size)
    
    # Variables that will be used to plot the losses from the discriminator and the adversarial models
    adversarial_loss = np.empty(shape=1)
    discriminator_loss = np.empty(shape=1)
    batches = np.empty(shape=1)
    
    
    print("Training Started .................")
    print(result.shape)
    # print(result.shape)
    for epoch in range(epochs):
            
            btch_ptr=0
            next_batch=0
            current_batch=batch_size
            start = time.time() 
            for batch_number in range(number_of_batches-1): 
                
                re_img = result[next_batch:current_batch]/255.0
                next_batch = current_batch
                current_batch = current_batch + batch_size
#                 print(re_img.shape)
#                 print(batch_number)
#                 print(next_batch)
#                 print(current_batch)
                
                noise = np.random.uniform(-1, 1, (batch_size, 100))
                fake_images = gen.predict(noise)
                
                real_y = np.ones(batch_size)
                fake_y = np.zeros(batch_size)
                
                # Let's train the discriminator
                d_loss = disc.train_on_batch(re_img, real_y)
                d_loss  = disc.train_on_batch(fake_images, fake_y) + d_loss
                discriminator_loss = np.append(discriminator_loss, d_loss)
                
                # Now it's time to train the generator
                noise = np.random.uniform(-1, 1, (batch_size, 100))
                # We try to mislead the discriminator by giving the opposite labels  
                fake_y = np.ones(batch_size)
                
                # Let's train the generator
                disc.trainable = False 
                g_loss=gan.train_on_batch(noise, fake_y)
                adversarial_loss=np.append(adversarial_loss , g_loss)  
                
                if (batch_number + 1) % 40 == 0: 
                    disc.trainable = True
                    gen.save('models2\\generator_epoch' + str(epoch) + '(' + str(batch_number) + ')' + '.hdf5')
                    disc.save('models2\\discriminator_epoch' + str(epoch) + '(' + str(batch_number) + ')' + '.hdf5')        
                    end = time.time()
                    time_elapsed = int(end-start)/60
                    print("BATCH NUMBER {}/{} ==================================================================================================== ".format(batch_number,number_of_batches-1))
                    print("EPOCH {} OF {} - ADVERSARIAL LOSS : {} , DISCRIMINATOR LOSS : {}, TIME ELAPSED : {} min.".format(epoch, epochs, g_loss, d_loss, time_elapsed))
                    print("\n")
```
