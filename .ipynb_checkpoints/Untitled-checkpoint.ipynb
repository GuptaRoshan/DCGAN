```python
from keras.models import load_model
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
noise = np.random.uniform(-1, 1, (25, 100))
```

```python
def plot_predicted_images(x):
    count=1;
    for val in x:
#         print(x)
        
        model = load_model("models2\\"+val)
        imgs =model.predict(noise)
        fig, axes = plt.subplots(nrows=2, ncols=8, figsize=(12,10))
        
        for indx, axis in enumerate(axes.flatten()):
            rnd_indx = np.random.randint(0, 25)
            imgplot = axis.imshow(imgs[rnd_indx], vmin=0 ,vmax=255,)
            axis.set_axis_off()         
        print("Generated images of {} epoch".format(count))
        plt.savefig("epoch0.png")
        plt.tight_layout(rect=[0, 0.08, 1.7, 0.95])
        count=count+1;
        print("\n")
    
```

```python
plot_predicted_images(["generator_epoch2(39).hdf5","generator_epoch2(39).hdf5"] )
# os.listdir("models")
```
